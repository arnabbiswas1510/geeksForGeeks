##Max and Second Max  
Given an array arr[] of size N of positive integers which may have duplicates. The task is to find maximum and second maximum from the array, and both of them should be distinct, so If no second max exists, then second max will be -1.

###MEDIUM

```python
def largestAndSecondLargest(arr):
    first=-10000
    second=-10000
    for num in arr:
        if num > first:
            second = first   # Remember this rotation is key
            first = num
        if num > second and num != first:   #Next key step to ensure than max and second are not the same
            second = num
    print ("largest is ", first)
    print ("second largest is ", second)

largestAndSecondLargest([12, 35, 1, 10, 34, 1])
```

##Reverse array in groups  
Given an array arr[] of positive integers of size N. Reverse every sub-array of K group elements.

Builds on reverse algo. Important to remember in **reverseInGroups** to consider mid and k when the loop oversteps the array

###MEDIUM

```python
def reverseInGroups(arr,k):
    n=len(arr)
    mid = int(k/2)
    cnt=0
    for i in range(0,n,k):      
        if(i+k) > n:        # Remember to have logic to adjust the end of array
            mid = int((n-i)/2)
            k=n-i
        while(cnt < mid):
            arr[i+cnt], arr[i+k-1-cnt] = arr[i+k-1-cnt], arr[i+cnt]
            cnt+=1
        cnt=0
    return arr
print(reverseInGroups([1,2,3,4,5,6],3))
```

##Equilibrium point  
Given an array A of N positive numbers. The task is to find the position where equilibrium first occurs in the array. Equilibrium position in an array is a position such that the sum of elements before it is equal to the sum of elements after it.

###EASY

```python
# O(n^2). Naive approach
def equilibriumPoint(A, N):
    ind=-1
    for i in range(N):
        left = right = 0        # Left and right must be initialized within outer loop
        for j in range(i):
            left=left+A[j]
        for j in range(i+1,N):
            right=right+A[j]
        if(left==right):
            ind=i
            break
    print("Equilibrium index:", ind)

arr = [1,3,5,2,2]
n = len(arr)
equilibriumPoint(arr, n)
```

###MEDIUM

```python
# O(n). Efficent approach
def equilibriumPoint(A, N):
    right=left=0
    for i in range(N):
        right=right+A[i]

    for i in range(N):
        right=right-A[i]
        if(left==right):
            break
        left=left+A[i]      #Remember to do this after the check else you will not find equilibrium
    print("Equilibrium index:", i)

arr = [1,3,5,2,2]
n = len(arr)
equilibriumPoint(arr, n)
```

##Rearrange an array with O(1) extra space  
Given an array arr[] of size N where every element is in range from 0 to n-1. Rearrange the given array so that arr`[i]` becomes arr`[arr[i]]`. This should be done with O(1) extra space.  

**Approach:** The array elements of the given array lies from 0 to n-1. Now an array element is needed that can store two different values at the same time. To achieve this increment every element at ith index is incremented by (arr`[arr[i]]` % n)*n.After the increment operation of first step, every element holds both old values and new values. Old value can be obtained by arr`[i]`%n and a new value can be obtained by arr`[i]`/n.

**How this can be achieved?**  
Let’s assume an element is a and another element is b, both the elements are less than n. So if an element a is incremented by b*n. So the element becomes a + b*n so when a + b*n is divided by n then the value is b and a + b*n % n is a.

###HARD

```python
def rearrange(arr, n):
    for i in range(n):
        arr[i] += (arr[arr[i]]%n)*n     # This step makes it store both numbers as described above, remeber you need
        # modulo so that you avoid indexerror
    for i in range(n):
        arr[i] = arr[i]//n             # The procedure to get the original number is modulo division
    print("Rearranged :",arr)


arr = [4, 0, 2, 1, 3]
n = len(arr)
rearrange(arr, n)
```

Method 3 (Efficient)
To solve this problem, we need to get two optimum indexes of arr[]: left index i and right index j. For an element arr`[i]`, we do not need to consider arr`[i]` for left index if there is an element smaller than arr`[i]` on left side of arr`[i]`. Similarly, if there is a greater element on right side of arr`[j]` then we do not need to consider this j for right index. So we construct two auxiliary arrays LMin[] and RMax[] such that LMin`[i]` holds the smallest element on left side of arr`[i]` including arr`[i]`, and RMax`[j]` holds the greatest element on right side of arr`[j]` including arr`[j]`. After constructing these two auxiliary arrays, we traverse both of these arrays from left to right. While traversing LMin[] and RMa[] if we see that LMin`[i]` is greater than RMax`[j]`, then we must move ahead in LMin[] (or do i++) because all elements on left of LMin`[i]` are greater than or equal to LMin`[i]`. Otherwise we must move ahead in RMax`[j]` to look for a greater j – i value.

This is a more complex variant of the trapping rainwater problem

###HARD

```python
#Efficent - O(nlogn)
def maxIndexDiff(arr, n):
    Lmin = [None]*n
    Rmax = [None]*n

    # Construct LMin[] such that
    # LMin[i] stores the minimum
    # value from (arr[0], arr[1],
    # ... arr[i])
    Lmin[0]=arr[0]
    for i in range(1,n):
        Lmin[i] = min(arr[i],Lmin[i-1])

    # Construct RMax[] such that
    # RMax[j] stores the maximum
    # value from (arr[j], arr[j+1],
    # ..arr[n-1])
    Rmax[n-1]=arr[n-1]
    for i in range(n-2,-1,-1):
        Rmax[i] = max(arr[i],Rmax[i+1])

    # Traverse both arrays from left
    # to right to find optimum j - i
    # This process is similar to
    # merge() of MergeSort
    # Lmin = <class 'list'>: [9, 2, 2, 2, 2, 2, 2, 2, 2, 0]
    # Rmax = <class 'list'>: [18, 18, 18, 18, 18, 18, 18, 18, 18, 0]
    i=j=0
    maxDiff=-1
    while i<n and j<n:
        if Lmin[i] < Rmax[j]:
            maxDiff=max(maxDiff,j-1)
            j+=1
        else:
            i+=1
    print("MaxIndexDiff:",maxDiff)

arr = [9, 2, 3, 4, 5, 6, 7, 8, 18, 0]
n = len(arr)
maxIndexDiff(arr, n)
```

##Trapping Rain Water

Given an array arr[] of N non-negative integers representing height of blocks at index i as Ai where the width of each block is 1. Compute how much water can be trapped in between blocks after raining.

The inefficient version of the below approach is similar but within the outer loop compute rmax(i) and lmax(i) similarly

###MEDIUM

```python
#Efficent - O(nlogn)
def findWater(arr, n):
    lmax=[None]*n
    rmax=[None]*n

    #Fill left array
    lmax[0]=arr[0]
    for i in range(1,n):
        lmax[i]=max(arr[i],lmax[i-1])

    #Fill right array
    rmax[n-1]=arr[n-1]
    for i in range(n-2,-1,-1):
        rmax[i]=max(arr[i],rmax[i+1])

    water=0
    for i in range(n):
        water += min(rmax[i],lmax[i])-arr[i]
    print("Trapped water: ",water)
    
arr = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
n = len(arr)
findWater(arr, n)
```

##Stock buy and sell

The cost of stock on each day is given in an array A[] of size N. Find all the days on which you buy and sell the stock so that in between those days your profit is maximum.

Efficient approach: If we are allowed to buy and sell only once, then we can use following algorithm. Maximum difference between two elements. Here we are allowed to buy and sell multiple times.  
Following is algorithm for this problem.

1.Find the local minima and store it as starting index. If not exists, return.  
2.Find the local maxima. and store it as ending index. If we reach the end, set the end as ending index.  
3.Update the solution (Increment count of buy sell pairs)  
4.Repeat the above steps if end is not reached.

###MEDIUM

```python
#Efficent - O(n), even though apparently seems to be n^2
def stockBuySell(price, n):
    if n<2:
        return
    # Use technique of nexted loop but not n^2 to print muliple buy/sell points
    i=0
    while i<n-1:
        # Find Local Minima
        # Note that the limit is (n-2) as we are
        # comparing present element to the next element
        while i<n-1 and price[i] > price[i+1]:
            i+=1
        # If we reached the end, break
        # as no further solution possible
        if i==n-2:
            break # No investment opportunity
        buy=i
        i+=1 #Remember to move to the next day here
        # Note that the limit is (n-1) as we are
        # comparing to previous element
        while i<n and price[i]>price[i-1]:
            i+=1
        sell = i-1 #Adjust here to prev day
        print("Buy on day: ", buy, "\t",
              "Sell on day: ", sell)

arr = [100, 180, 260, 310, 40, 535, 695]
n = len(arr)
stockBuySell(arr, n)
```

##Largest Sum Contiguous Subarray

Write an efficient program to find the sum of contiguous subarray within a one-dimensional array of numbers which has the largest sum.  

###MEDIUM


```python
def kadane(arr, n):
    import sys
    fullMax=-sys.maxsize
    currMax=-sys.maxsize
    for i in range(n):
        currMax = max(arr[i], currMax + arr[i])
        fullMax = max(fullMax, currMax)
    return fullMax

arr = [-2, -3, 4, -1, -2, 1, 5, -3]
n = len(arr)
print("Max contiguous subsequence", kadane(arr, n))
```

Remember that fullMax is set before currMax is reset

```python
def kadaneIndex(arr, n):
    import sys
    fullMax=-sys.maxsize
    currMax=-sys.maxsize
    s = end =0
    for i in range(n):
        currMax += arr[i]
        if currMax > fullMax:
            fullMax=currMax
            start = s               # Note that start and s are two separate variables
            end = i
        if currMax < 0:
            currMax = 0
            s = i+1
    return fullMax, start, end

arr = [-2, -3, 4, -1, -2, 1, 5, -3]
n = len(arr)
print("Max contiguous subsequence: %d, start: %d, end: %d" %(kadaneIndex(arr, n)))
```

##Max Circular Subarray Sum

Given an array arr[] of N integers arranged in a circular fashion. Your task is to find the maximum contigious subarray sum.

###MEDIUM

```python
def kadaneCircular(arr, n):
    sum=0
    # Case 1: get the maximum sum using standard kadane's
    # algorithm
    kad=kadane(arr,n)
    # Case 2: Now find the maximum sum that includes corner
    # elements.
    for i in range(n):
        sum+=arr[i]
        arr[i]=-arr[i]
    # Max sum with corner elements will be:
    # array-sum - (-max subarray sum of inverted array)
    kadInv=sum+kadane(arr,n)
    # The maximum circular sum will be maximum of two sums
    return max(kad,kadInv)


arr = [11, 10, -20, 5, -3, -5, 8, -13, 10]
n = len(arr)
print("Max contiguous subsequence", kadaneCircular(arr, n))
```
